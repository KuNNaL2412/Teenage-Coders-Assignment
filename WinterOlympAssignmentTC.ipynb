Q2. Number of Gold Medals wins in entire Winter Olympics


```python
for name in df['Medal'].unique():
  s = 0
  for i in df['Medal']:
    if (i == name):
      s += 1
  print(name,":", s)
```

Q3. Which country is having most numbers of Medals


```python
most_medals = df.groupby(['Country'])['Medal'].sum().sort_values(ascending = False).head(1)
most_medals
```

Q4. Which country is having most numbers of Gold Medals

```python
goldMedals = df[(df.Medal == 'Gold')]
```

```python
plt.figure(figsize=(20, 10))
plt.title('Distribution of Gold Medals')
sns.countplot(goldMedals['Country'])
plt.show()
```

Q5. Which sport is having most number of Gold Medals


```python
plt.figure(figsize=(20, 10))
plt.title('Distribution of Gold Medals')
sns.countplot(goldMedals['Sport'])
plt.show()
```

Q6. Top 5 Disciplines in terms of Medals


```python
me = df.groupby(['Discipline'])['Medal'].sum().sort_values(ascending = False).head(5)
me
```

Q7. Top 5 Disciplines in terms of Bronze Medals


```python
bronzeMedals = df[(df.Medal == 'Bronze')]
plt.figure(figsize=(20, 10))
plt.title('Distribution of Bronze Medals')
sns.countplot(bronzeMedals['Discipline'])
plt.show()
```
